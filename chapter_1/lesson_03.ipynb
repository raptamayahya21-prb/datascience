# Perbedaan / overview matplotlib dan searborn

## Matplotlib
- tujuan: kontrol penuh, fleksibel
- level: low level
- input data: array numpy, list
- sintaks: lebih panjang 
- aturan praktis: gunakan untuk kostumisasi mendalam

## Searborn
- tujuan: analisis statistik & estetik
- level: high level
- input data: pandas framework
- sintaks: lebih ringkas
- aturan praktis: gunakan untuk ekplorasi cepat

1. Matplotlib

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
data = {
    'pengeluaran_iklan' : [10, 15, 12, 18, 22],
    'penjualan' : [100, 140, 110, 180, 210]
}

df = pd.DataFrame(data)
print(df)
```

```python
# membuat plot

plt.figure(figsize=(8,5))
plt.scatter(df['pengeluaran_iklan'],df['penjualan'])
plt.title('Hubungan Iklan dan Penjualan (Matplotlib)')
plt.xlabel('Pengeluaran Iklan (Rp. 1000)')
plt.ylabel('Penjualan (Rp. 1000)')
plt.grid(True)
plt.show()
```

2. Seaborn

```python
plt.figure(figsize=(8,5))
sns.scatterplot(data=df, x='pengeluaran_iklan', y='penjualan')
plt.title('Hubungan Iklan dan Penjualan (Matplotlib)')
plt.xlabel('Pengeluaran Iklan (Rp. 1000)')
plt.ylabel('Penjualan (Rp. 1000)')
plt.show()
```
