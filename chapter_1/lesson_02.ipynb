```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
#siapkan data sederhana 
x = np.linspace(0, 10, 100)
y = np.sin(x)
```

```python
sns.set_style("whitegrid")
```

```python
plt.figure(figsize=(8,5))
plt.plot(x,y)
plt.title("tes plot sinus sederhana")
plt.xlabel("nilai X")
plt.ylabel("nilai sin(x)")
plt.show()
```
