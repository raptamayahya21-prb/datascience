# Bar CHART

```python
# Langkah 1: import Matplotlib
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Langkah 2: Siapkan Data
# Sumbu X: merk laptop
produk = ["apple", "acer", "hp", "lenovo", "dell", "asus"]
# Sumbu Y: jumlah terjual
terjual = [10, 20, 5, 30, 5, 60]
```

```python
# Langkah 3: Siapkam Bar Chart
plt.figure(figsize=(4,3))
plt.bar(produk,terjual, color='r')
plt.show()
```

```python
# Langkah 4: Bar Chart Multi Warna
plt.figure(figsize=(4,3))
warna = ["#fd0000", "#FF8000", "#0000FF", "#111111", "#FFF70D", "#44cc55"] #warna warni
plt.bar(produk,terjual, color=warna, edgecolor='black')
plt.show()
```

```python
# Langkah 5: Bar Chart Horizontal
plt.figure(figsize=(4,3))
warna = ["#fd0000", "#FF8000", "#0000FF", "#111111", "#FFF70D", "#44cc55"] #warna warni
plt.barh(produk,terjual, color=warna, edgecolor='black')
plt.show()
```

```python
plt.figure(figsize=(4,3))
plt.barh(produk,terjual, color=warna, edgecolor='black')
plt.gca().invert_yaxis()  # Membalik sumbu Y agar produk dengan penjualan tertinggi di atas
plt.show()
```

## Contoh Study kasus

```python
# Langkah 2: Siapkan Data
# Sumbu X: merk laptop
produk = ["apple", "acer", "hp", "lenovo", "dell", "asus"]
# Sumbu Y: jumlah terjual
terjual_2024 = [10, 20, 5, 30, 5, 60]
terjual_2025 = [15, 25, 10, 10, 2, 40]
```

```python
# menentukan posisi untuk setiap group
x = np.arange(len(produk))
lebar_batang = 0.35 # lebar setiap batang
```

```python
plt.figure(figsize=(5,5))
ax = plt.gca()

# membuat batang untuk data 2023
batang1 = ax.bar(x - lebar_batang/2, terjual_2024, lebar_batang, color="#44cc55")
batang2 = ax.bar(x + lebar_batang/2, terjual_2025, lebar_batang, color="#0055ff")

# mengatur label sumbu X
ax.set_xticks(x)
ax.set_xticklabels(produk)

# menambah label angka di setiap batang
ax.bar_label(batang1, padding=3)
ax.bar_label(batang2, padding=3)

plt.show()
```
