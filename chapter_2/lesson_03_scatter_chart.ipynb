```python
# Langkah 1: import library matplotlib
import matplotlib.pyplot as plt
```

```python
# Langkah 2: Siapkan data
# Sumbu X: Pengeluaran Iklan (dalam jutaan rupiah)
Pengeluaran_iklan = [2.5, 3.0, 3.2, 4.5, 5.0, 5.5, 6.0, 6.2, 7.0, 8.0]
# Sumbu Y: Penjualan
Penjualan = [150, 180, 190, 250, 280, 310, 340, 330, 380, 420]
```

```python
# Langkah 3: Membuat figure dan Scatter Plot
plt.figure(figsize=(4,4))
plt.scatter(Pengeluaran_iklan, Penjualan)
plt.show()
```

```python
# Langkah 4: Menambahkan Style
plt.figure(figsize=(4,4))
plt.scatter(Pengeluaran_iklan, Penjualan,
            color="green",
            s=100,
            alpha=0.6,
            edgecolor="black",
            marker="o",
            )
plt.show()
```

```python
import numpy as np
Penjualan = [150, 180, 190, 250, 280, 310, 340, 330, 380, 420]
warna = 1 + 0.1*np.arange(len(Penjualan)) # membuat array gradasi warna
```

```python
# Langkah 5: Menambahkan Gradasi Warna
plt.figure(figsize=(4,4))
plt.scatter(Pengeluaran_iklan, Penjualan,
            c=warna,
            cmap='viridis',
            s=100,
            alpha=0.6,
            edgecolor="black",
            marker="o",
            )

# Menghitung regresi
# Mengguanakan polyfit dari numpy untuk regresi, derajat 1 (linear)
beta_1, beta_0 = np.polyfit(Pengeluaran_iklan, Penjualan, 1)

# membuat garis 
data_x = np.array(Pengeluaran_iklan)
data_y = beta_1 * data_x + beta_0
plt.plot(data_x, data_y, color='green', linestyle='--', linewidth=2)
 
plt.show()
```

```python
print("beta_1", beta_1)
print("beta_0", beta_0)
```
