# Histogram

histogram adalah visualisasi yang penting untuk memahami distribusi.menggunakan interval atau "bins" untuk membagi data kita dan memberikan frekuensi pada setiap interval

"pada rentang usia berapa pelanggan kami paling banyak?"

### 1. Histogram Dasar

```python
# Langkah 1
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Langkah 2
# membuat 100 nilai anak yang terdistribusi normal dengan rata-rata 75,
# dan standard deviasi 10

np.random.seed(42) # kalau pakai ini nilainya bisa berubah
nilai_ujian = np.random.normal(loc=75, scale=10, size=100)
print(nilai_ujian)
```

```python
# Langkah 3: Membuat Histogram
plt.figure(figsize=(6,6))

plt.hist(nilai_ujian)

plt.show()
```

### 2. Mengontrol Interval

```python
# Langkah 3: Membuat Histogram
plt.figure(figsize=(6,6))

plt.hist(nilai_ujian, bins=15) #bins / interval

plt.show()
```

### 3. Mengubah Style

```python
# Langkah 5: Membuat Histogram dengan style
plt.figure(figsize=(6,6))

plt.hist(nilai_ujian,
        bins=15, # bins / interval
        color="green",
        edgecolor="black",
        alpha=0.7,
        ) 

plt.grid(axis='y', linestyle='--', alpha=0.5)

# menambahkan nilai rata-rata
plt.axvline(nilai_ujian.mean(), color='blue', linestyle=':', linewidth=2)

plt.show()
```

### 4. Membandingkan distribusi

```python
np.random.seed(0)
nilai_ujian_1 = np.random.normal(loc=75, scale=8, size=100)
nilai_ujian_2 = np.random.normal(loc=72, scale=12, size=100)

plt.figure(figsize=(6,6))

plt.hist(nilai_ujian_1,
        bins=12, # bins / interval
        color="green",
        edgecolor="black",
        alpha=0.5,
        label="Kelas A"
        ) 

plt.hist(nilai_ujian_2,
        bins=12, # bins / interval
        color="brown",
        edgecolor="black",
        alpha=0.5,
        label="Kelas B"
        ) 

plt.grid(axis='y', linestyle='--', alpha=0.5)

# menambahkan nilai rata-rata
plt.axvline(nilai_ujian.mean(), color='blue', linestyle=':', linewidth=2)

plt.show()
```
