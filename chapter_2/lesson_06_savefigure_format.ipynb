```python
import matplotlib.pyplot as plt

bulan = ["jan", "feb", "mar", "apr", "mei", "jun"]
penjualan_A = [200, 220, 250, 240, 280, 300]
penjualan_B = [180, 190, 210, 230, 250, 260]

# pola dasar
plt.figure(figsize=(8,4))
plt.plot(bulan, penjualan_A, marker='o', label='Gadget A')
plt.plot(bulan, penjualan_B, marker='s', label='Gadget B')
plt.title(
    "Perbandingan Penjualan Gadget A dan Gadget B",
    fontsize=14,
    fontweight='bold',
    color='darkblue',
    loc='right'
)
plt.xlabel(
    "Bulan",
    fontsize=12,
    fontweight='bold',
    color='darkgreen',
    loc='center'
)
plt.ylabel(
    "Unit Terjual",
    fontsize=12,
    fontweight='bold',
    color='darkgreen',
    loc='center'
)
plt.legend(
    loc='lower right',
    ncol=2,
    title='Produk',
    shadow=True
)

plt.grid(True, linestyle='--', alpha=0.5)

# 1. saving plot (harus dilakukan sebelum plt.show())

plt.savefig('plot1.png', transparent=True) #raster
plt.savefig('plot2.jpg', dpi=600) #raster
plt.savefig('plot3.pdf') #vector
plt.savefig('plot4.svg') #vector
plt.savefig('plot5.jpg', dpi=600, bbox_inches="tight") #raster
plt.show()

```
