Untuk membuat garis grafik,kita bisa gunakan plt.plot()

1. Garis Grafik Paling Dasar

```python
# Langkah 1: Import Library Matplotlib

import matplotlib.pyplot as plt
```

```python
# Langkah 2: Siapkan Data Kita
# Sumbu X: Bulan
bulan = ['Jan', 'Feb', 'Mar', 'Apr', 'Mei', 'Jun']
# Sumbu Y: Suhu Dalam Celcius
suhu_celcius = [27, 28, 29, 31, 30, 29]
```

```python
# Langkah 3: Buat Plot
plt.plot(bulan, suhu_celcius)
plt.show()
```

```python
# Langkah 4: Buat Plot dengan warna berbeda
plt.plot(bulan, suhu_celcius, color='g')
plt.show()
```

```python
# Langkah 5: Buat Plot dengan warna berbeda dan style berbeda
plt.plot(bulan, suhu_celcius, color='g', linestyle='--')
plt.show()
```

```python
# Langkah 6: Buat Plot dengan warna berbeda dan style berbeda, dan marker
plt.plot(bulan, suhu_celcius, color='g', linestyle='--', marker='o')
plt.show()
```

```python
# Cara penggunaan line chart
# Sumbu X: Bulan
bulan = ['Jan', 'Feb', 'Mar', 'Apr', 'Mei', 'Jun']
# Sumbu Y: Suhu Dalam Celcius
suhu_jakarta = [27, 28, 29, 31, 30, 29]
suhu_bandung = [22, 23, 24, 27, 26, 25]
```

```python
plt.plot(bulan,suhu_jakarta, marker="o", color="r", linestyle="-")
plt.plot(bulan,suhu_bandung, marker="s", color="b", linestyle="--")
plt.show()
```
