### Kostumisasi title, tabel, legend

```python
import matplotlib.pyplot as plt

# data dasar

bulan = ["jan", "feb", "mar", "apr", "mei", "jun"]
penjualan_A = [200, 220, 250, 240, 280, 300]
penjualan_B = [180, 190, 210, 230, 250, 260]

# pola dasar
plt.figure(figsize=(8,4))
plt.plot(bulan, penjualan_A, marker='o', label='Gadget A')
plt.plot(bulan, penjualan_B, marker='s', label='Gadget B')
plt.show()
```

```python
# 1.Menambahkan dan menyesuaikan judul

plt.figure(figsize=(8,4))
plt.plot(bulan, penjualan_A, marker='o', label='Gadget A')
plt.plot(bulan, penjualan_B, marker='s', label='Gadget B')
plt.title("Perbandingan Penjualan Gadget A dan Gadget B")
plt.show()
```

```python
# 2.Menyesuaikan Judul

plt.figure(figsize=(8,4))
plt.plot(bulan, penjualan_A, marker='o', label='Gadget A')
plt.plot(bulan, penjualan_B, marker='s', label='Gadget B')
plt.title(
    "Perbandingan Penjualan Gadget A dan Gadget B",
    fontsize=14,
    fontweight='bold',
    color='darkblue',
    loc='right'
)
plt.show()
```

```python
# 3. Menambahkan label

plt.figure(figsize=(8,4))
plt.plot(bulan, penjualan_A, marker='o', label='Gadget A')
plt.plot(bulan, penjualan_B, marker='s', label='Gadget B')
plt.title(
    "Perbandingan Penjualan Gadget A dan Gadget B",
    fontsize=14,
    fontweight='bold',
    color='darkblue',
    loc='right'
)
plt.xlabel("Bulan")
plt.ylabel("Unit Terjual")
plt.show()
```

```python
# 4. Mengkostumasi label

plt.figure(figsize=(8,4))
plt.plot(bulan, penjualan_A, marker='o', label='Gadget A')
plt.plot(bulan, penjualan_B, marker='s', label='Gadget B')
plt.title(
    "Perbandingan Penjualan Gadget A dan Gadget B",
    fontsize=14,
    fontweight='bold',
    color='darkblue',
    loc='right'
)
plt.xlabel(
    "Bulan",
    fontsize=12,
    fontweight='bold',
    color='darkgreen',
    loc='center'
)
plt.ylabel(
    "Unit Terjual",
    fontsize=12,
    fontweight='bold',
    color='darkgreen',
    loc='center'
)
plt.show()
```

```python
# 5. Menambahkan legend

plt.figure(figsize=(8,4))
plt.plot(bulan, penjualan_A, marker='o', label='Gadget A')
plt.plot(bulan, penjualan_B, marker='s', label='Gadget B')
plt.title(
    "Perbandingan Penjualan Gadget A dan Gadget B",
    fontsize=14,
    fontweight='bold',
    color='darkblue',
    loc='right'
)
plt.xlabel(
    "Bulan",
    fontsize=12,
    fontweight='bold',
    color='darkgreen',
    loc='center'
)
plt.ylabel(
    "Unit Terjual",
    fontsize=12,
    fontweight='bold',
    color='darkgreen',
    loc='center'
)
plt.legend()
plt.show()
```

```python
# 6. Mengkostumasi legend

plt.figure(figsize=(8,4))
plt.plot(bulan, penjualan_A, marker='o', label='Gadget A')
plt.plot(bulan, penjualan_B, marker='s', label='Gadget B')
plt.title(
    "Perbandingan Penjualan Gadget A dan Gadget B",
    fontsize=14,
    fontweight='bold',
    color='darkblue',
    loc='right'
)
plt.xlabel(
    "Bulan",
    fontsize=12,
    fontweight='bold',
    color='darkgreen',
    loc='center'
)
plt.ylabel(
    "Unit Terjual",
    fontsize=12,
    fontweight='bold',
    color='darkgreen',
    loc='center'
)
plt.legend(
    loc='lower right',
    ncol=2,
    title='Produk',
    shadow=True
)

plt.grid(True, linestyle='--', alpha=0.5)
plt.show()
```
