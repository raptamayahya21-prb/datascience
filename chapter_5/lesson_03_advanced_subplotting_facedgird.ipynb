#  Catplot

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
tips_df = sns.load_dataset('tips')
print(tips_df.head())
sns.set_theme(style="whitegrid")
```

```python
# menggunakan catplot
sns.catplot(
    data=tips_df,
    kind="box",
    x="day",
    y="total_bill",
    col="smoker"   
)
plt.show()
```

### 3. FacetGrid

Grid yang terstruktur:
- data
- row
- col
- hue


```python
# Langkah 1: inisialisasi Grid

g = sns.FacetGrid(
    data=tips_df,
    row="time",
    col="smoker",
    hue="sex",
    margin_titles=True
)
# Langkah 2: petakan fungsi scatterplot
g.map(sns.scatterplot, "total_bill","tip", alpha=0.7)
plt.show()
```

```python
# Langkah 1: inisialisasi Grid

g = sns.FacetGrid(
    data=tips_df,
    row="time",
    col="smoker",
    hue="sex",
    margin_titles=True
)
# Langkah 2: petakan fungsi scatterplot
g.map(sns.scatterplot, "total_bill","tip", alpha=0.7)

# Langkah 3: styling
g.add_legend()
g.set_axis_labels("total_tagihan","tip")
g.set_titles(col_template="{col_name}", row_template="{row_name}")
plt.show()
```
