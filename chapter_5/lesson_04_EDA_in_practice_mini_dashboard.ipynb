# Studi kasus: Memahami Faktor Harga Berlian
## Tujuan Bisnis:
- sebuah perusahaan perhiasan ingin memahami faktor-faktor apa saja yang paling memengaruhi harga berlian. Mereka memberi "dataset" untuk dianalisis.

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
# data
diamonds_tf = sns.load_dataset('diamonds')
print(diamonds_tf.head())
```

## Langkah 1: Gambaran Umum dengan pairplot

```python
# mengambil sampel acak 500 baris untuk pairplot
sampel_df = diamonds_tf.sample(n=500, random_state=42)

# membuat pairplot dengan hue berdasarkan kualitas potongan ('cut')
sns.pairplot(sampel_df, hue='cut', palette='viridis')

plt.show()
```

## Langkah 2: mengukur korelasi dengan Heatmap

```python
# menghitung matriks korelasi
correlation_matrix = diamonds_tf.corr(numeric_only=True)

# membuat Heatmap
plt.figure(figsize=(8,6))
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm", fmt='.2f')
plt.title("matriks korelasi fitur bearlian", fontsize=16)
plt.show()
```

## Langkah 3: Menjawab pertanyaan spesifik
bagaimana hubungan antara karat dan price berubah tergantung pada kualitas dua potongan dan dua berlian

```python
# relplot

sns.relplot(
    data=sampel_df,
    x="carat",
    y="price",
    col='cut',
    hue='color',
    col_wrap=3,
    palette='YlOrBr'
)

plt.suptitle("Hubungan carat vs price berdasarkan kualitas potongan dan warna", y=1.02)
plt.show()
```
