EDA - Exploratory Data Analysis memahami dataset. untuk menjawab pertanyaan:

- variabel mana yang tampaknya saling berhubungan
- bagaimana distribusi umum dari setiap varriabel?

dengan seaborn: pairplot dan joinplot


### 1. Pairplot

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# mengambil dataset penguins
penguins_df = sns.load_dataset('penguins')
print(penguins_df.head())
```

```python
plt.figure(figsize=(8,4))
sns.pairplot(data=penguins_df, hue='species')
plt.show()
```

```python
sns.pairplot(data=penguins_df, hue='island')
plt.show()
```

### 2. Joinplot

fokus pada hubungan dua variabel

```python
tips_df = sns.load_dataset('tips')
sns.set_theme(style="whitegrid")
sns.jointplot(data=tips_df, x='total_bill', y='tip') 
plt.show()
```

```python
sns.jointplot(data=tips_df, x='total_bill', y='tip', kind='reg') 
plt.show()
```

```python
sns.jointplot(data=tips_df, x='total_bill', y='tip', kind='hex') 
plt.show()
```
