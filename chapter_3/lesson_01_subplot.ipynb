# Subplot

```python
# Langkah 1: mambuat library

import matplotlib.pyplot as plt
```

```python
# Langkah 2:Data
bulan = ["jan", "feb", "mar", "apr", "mey", "jun"]
suhu = [27,28,29,31,30,29]
curah_hujan = [300,250,200,150,100,80]
```

```python
#langkah 3: membuat plot
plt.figure(figsize=(8,4))

# plot pertama
plt.subplot(2,1,1) #row,col,index
plt.plot(bulan,suhu)
plt.title("Suhu Hujan Rata-rata Bulanan")
plt.ylabel("C")

# plot kedua
plt.subplot(2,1,2) #row,col,index
plt.plot(bulan,curah_hujan)
plt.title("Curah Hujan Rata-rata Bulanan")
plt.ylabel("mm")

plt.tight_layout()
plt.show()
```

# Subplot

```python
# subplots
fig, ax = plt.subplots(nrows=2, ncols=1, figsize=(8,8))

# ambil axis pertama
ax[0].plot(bulan, suhu)
ax[0].set_title("Suhu Hujan Rata-rata Bulanan")
# ambil axis kedua
ax[1].plot(bulan, curah_hujan)
ax[1].set_title("Curah Hujan Rata_rata Bulanan")
```

```python
import numpy as np
# subplots
fig, ax = plt.subplots(nrows=2, ncols=2, figsize=(8,8), sharex=True)

# ambil axis pertama
ax[0,0].plot(bulan, suhu)
ax[0,0].set_title("Suhu Hujan Rata-rata Bulanan")

# ambil axis kedua
ax[0,1].plot(bulan, curah_hujan)
ax[0,1].set_title("Curah Hujan Rata_rata Bulanan")

# data baru
penjualan = [200,150,200,100,160,180]
kepuasan_pelanggan = np.random.randint(70,100,6)

# ambil axis ketiga
ax[1,0].bar(bulan, penjualan)
ax[1,0].set_title("Penjualan Rata-rata Bulanan")

# ambil axis keempat
ax[1,1].scatter(bulan, kepuasan_pelanggan)
ax[1,1].set_title("Kepuasan Pelanggan Rata-rata Bulanan")

# judul
fig.suptitle("Dashboard Bulanan")
plt.tight_layout()

plt.show()
```
