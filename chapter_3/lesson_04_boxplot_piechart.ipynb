### Boxplot
- nilai maksimum
- nilai minimum
- Q1,Q2,Q3
- outliers

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# membuat data sampel
np.random.seed(10)
pengeluaran_jakarta = np.random.normal(100,20,200)
pengeluaran_bandung = np.random.normal(80,30,200)
pengeluaran_surabaya = np.random.normal(90,25,200)
```

```python
# menggabungkan sebuah data ke dalam list
data = [pengeluaran_jakarta, pengeluaran_bandung, pengeluaran_surabaya]
```

```python
# figure dan axes
fig, ax= plt.subplots(figsize=(8,4))

# membuat boxplot
bp = ax.boxplot(data, patch_artist=True, notch=True)

colors = ['lightblue', 'lightgreen', 'lightpink'] 
# bp['boxes'][0].set_facecolor('lightblue')
# bp['boxes'][1].set_facecolor('lightgreen')
# bp['boxes'][2].set_facecolor('lightpink')

for patch, color in zip(bp['boxes'],colors): # looping
    patch.set_facecolor(color)
```

### Pie Chart


```python
# Data
sistem_operasi = ['android', 'ios', 'kaios', 'harmonyos']
pangsa_pasar = [70,25,2,3]
```

```python
fig, ax = plt.subplots(figsize=(8,4))

explode = (0, 0.1, 0, 0)

ax. pie(
    pangsa_pasar,
    labels=sistem_operasi,
    autopct='%1.1f%%',
    explode= explode,
    colors= ['green', 'blue', 'white', 'red'],
    startangle = 145
)

plt.show()
```
