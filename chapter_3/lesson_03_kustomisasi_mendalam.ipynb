### Mengatur batas suhu

```python
import matplotlib.pyplot as plt
```

```python
# Data
bulan_angka = [1,2,3,4,5,6,7,8]
penjualan = [200,220,250,240,280,300,310,330]
```

```python
fig, ax = plt.subplots(figsize=(8,4))

ax.plot(bulan_angka,penjualan,marker='o')
ax.set_title("Fokus Pada Penjualan Kuartal Kedua")
ax.set_xlabel("Bulan(angka)")
ax.set_ylabel("Unit Terjual")

# batas sumbu y
ax.set_ylim(220,320)

# batas sumbu x
ax.set_xlim(3.5, 6.5)

ax.grid(True)
plt.show()
```

### Menyesuaikan Tiks

```python
bulan = ["jan","feb","mar","apr","mei","jun","jul","agus"]

fig, ax = plt.subplots(figsize=(8,4))

ax.plot(bulan_angka,penjualan,marker='o')
ax.set_title("Penjualan Dengan Label Bulan Yang Disesuaikan")

# Mengatur Posisi Tiks Agar Sesuai
ax.set_xticks(bulan_angka)

ax.set_xticklabels(bulan, rotation=45, ha='right')

plt.show()
```

### Menambah Anotasi dan Teks

```python
fig, ax = plt.subplots(figsize=(8,4))

ax.plot(bulan,penjualan,marker='o')
ax.set_title("Fokus Pada Penjualan Kuartal Kedua")
ax.set_xlabel("Bulan(angka)")
ax.set_ylabel("Unit Terjual")

# anotasi
ax.annotate(
    'penurunan tak terduga',
    xy=('apr', 240),
    xytext=('feb', 270),
    arrowprops=dict(
        facecolor='black',
        arrowstyle='->', # cari tipe panah apa aja 
        connectionstyle='arc3,rad=0.2'
    ),
    fontsize=12,
)

ax.grid(True)
plt.show()
```

### Anotasi Persamaan

```python
import numpy as np

bulan_angka = np.array([1,2,3,4,5,6,7,8])
penjualan = np.array([200,220,250,240,280,300,310,330])
fig, ax = plt.subplots(figsize=(8,4))

ax.scatter(bulan_angka,penjualan,marker='o')
ax.set_title("Fokus Pada Penjualan Kuartal Kedua")
ax.set_xlabel("Bulan(angka)")
ax.set_ylabel("Unit Terjual")

m, c = np.polyfit(bulan_angka, penjualan, 1)

ax.plot(bulan_angka, m*bulan_angka + c, color='red')
persamaan_text = r"$ \hat{y} $"f"$ ={m:.2f}x + {c:.2f}$" # harus difahami
ax.text(4,320, persamaan_text, fontsize=14, bbox=dict(facecolor='white', alpha=0.5))

plt.grid(True)
plt.show()
```
