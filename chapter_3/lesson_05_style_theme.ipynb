### Style sama Tema

```python
import matplotlib.pyplot as plt
import numpy as np
print(plt.style.available)
```

```python
x = np.linspace(0,10,100)
y1 = np.sin(x) 
y2 = np.cos(x)
```

```python
plt.figure(figsize=(8,4))
plt.plot(x,y1,label="sin(x)")
plt.plot(x,y2,label="cos(x)")
plt.title('plot style default matplotlib')
plt.xlabel('nilai x')
plt.ylabel('nilai y')
plt.legend()
plt.show()
```

### GGPLOT

```python
#ngambil
plt.style.use('ggplot')

#plot
plt.figure(figsize=(8,4))
plt.plot(x,y1,label="sin(x)")
plt.plot(x,y2,label="cos(x)")
plt.title('plot style ggplot matplotlib')
plt.xlabel('nilai x')
plt.ylabel('nilai y')
plt.legend()
plt.show()
```

```python
#ngambil
plt.style.use('fivethirtyeight')

#plot
plt.figure(figsize=(8,4))
plt.plot(x,y1,label="sin(x)")
plt.plot(x,y2,label="cos(x)")
plt.title('plot style 538 matplotlib')
plt.xlabel('nilai x')
plt.ylabel('nilai y')
plt.legend()
plt.show()
```

```python
#ngambil
plt.style.use('dark_background')

#plot
plt.figure(figsize=(8,4))
plt.plot(x,y1,label="sin(x)")
plt.plot(x,y2,label="cos(x)")
plt.title('plot style dark_background matplotlib')
plt.xlabel('nilai x')
plt.ylabel('nilai y')
plt.legend()
plt.show()
```

```python
# plot 1
plt.style.use('default')
plt.figure(figsize=(8,4))
plt.plot(x,y1,label="sin(x)")
plt.plot(x,y2,label="cos(x)")
plt.title('plot style default matplotlib')
plt.xlabel('nilai x')
plt.ylabel('nilai y')
plt.legend()
plt.show()

# plot 2
with plt.style.context('seaborn-v0_8-bright'):
    plt.figure(figsize=(8,4))
    plt.plot(x,y1,label="sin(x)")
    plt.plot(x,y2,label="cos(x)")
    plt.title('plot style seaborn-v0_8-bright matplotlib')
    plt.xlabel('nilai x')
    plt.ylabel('nilai y')
    plt.legend()
plt.show()

# plot 3
plt.style.use('default')
plt.figure(figsize=(8,4))
plt.plot(x,y1,label="sin(x)")
plt.plot(x,y2,label="cos(x)")
plt.title('plot style default matplotlib')
plt.xlabel('nilai x')
plt.ylabel('nilai y')
plt.legend()
plt.show()
```
