## Plotting Pendekatan Object oriented

```python
# sebelumnya

import matplotlib.pyplot as plt

bulan = ["jan", "feb", "mar", "apr"]
penjualan = [100,200,150,50]

fig, ax = plt.subplots(nrows=2,ncols=2,figsize=(8,8))

fig.suptitle("Grafik Perusahaan")
fig.supxlabel("2025")
fig.supylabel("Laptop")


ax[0,0].plot(bulan, penjualan, label="unit")
ax[0,0].set_title("penjualan")
ax[0,0].set_xlabel("Bulan")
ax[0,0].set_ylabel("Jumlah")
ax[0,0].legend()

plt.tight_layout()
plt.show()
```
