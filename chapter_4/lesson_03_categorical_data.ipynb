### Categorical Data

1. Barplot 

```python
# masukkan library
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
```

```python
#data
tips_df = sns.load_dataset('tips')
sns.set_theme(style='whitegrid')
print(tips_df.head())
```

```python
#membuat barplot
plt.figure(figsize=(8,4))
sns.barplot(data=tips_df,x="day",y='total_bill')

plt.title("Rata-rata Total Tagihan Per hari")
plt.xlabel( "Hari")
plt.ylabel("Rata-rata Tagihan, ratus Rp.")
plt.show()
```

```python
plt.figure(figsize=(8,4))
sns.barplot(data=tips_df,x="day",y='total_bill', estimator=np.median, errorbar=None)

plt.title("Rata-rata Total Tagihan Per hari")
plt.xlabel( "Hari")
plt.ylabel("Rata-rata Tagihan, ratus Rp.")
plt.show()
```

2. CountPlot

```python
# membuat countplot
plt.figure(figsize=(8,4))
sns.countplot(data=tips_df,x='day', hue='time')

plt.title("Jumlah Pelanggan Per Hari")
plt.xlabel("Hari")
plt.ylabel("Jumlah")
plt.show()
```
