### 1. Visualizing Relationship


1. Scatter Plot

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
tips_df = sns.load_dataset('tips')
print(tips_df.head())
```

```python
# membuat scatter plot
sns.set_theme(style='whitegrid')
plt.figure(figsize=(8,4))
sns.scatterplot(data=tips_df,x='total_bill',y='tip', hue='sex', style='smoker')
plt.title("Hubungan total tagihan dan bill")
plt.show()
```

2. Line Plot

```python
# dataset 'fmri'
fmri_df= sns.load_dataset('fmri')
print(fmri_df.head())
```

```python
# membuat line plot
plt.figure(figsize=(8,4))
sns.lineplot(data=fmri_df,x='timepoint',y='signal',hue='event')
plt.title("Perkembangan sinyal otak dari waktu ke waktu")
plt.show()

plt.figure(figsize=(8,4))
sns.scatterplot(data=fmri_df,x='timepoint',y='signal',hue='event')
plt.title("Perkembangan sinyal otak dari waktu ke waktu")
plt.show()
```

3. Relplot

```python
plt.figure(figsize=(8,4))
sns.relplot(data=tips_df,x='total_bill',y='tip',col='day')
plt.show()
```
