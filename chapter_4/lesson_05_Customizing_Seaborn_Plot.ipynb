### 2. Mengatur Tema Seaborn

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
tips_df = sns.load_dataset('tips')
print(tips_df.head())

sns.set_theme(style="white")
sns.boxplot(data=tips_df, x='day', y='total_bill')
plt.show()
```

```python
tips_df = sns.load_dataset('tips')

# context: notebook, paper, talk, poster.
sns.set_theme(style="white", context='paper')
# sns.set_style("white")
# sns.set_context("notebook")
sns.boxplot(data=tips_df, x='day', y='total_bill')
plt.show()
```

```python
sns.set_theme(style="white", context="paper")
sns.boxplot(data=tips_df, x='day', y='total_bill', hue='sex', palette='viridis')
plt.title("Distribusi Tagihan Per Hari")
plt.show()
```

### 3. Kustom detail dengan objek Matplotlib

```python
plt.figure(figsize=(8,4))

ax = sns.scatterplot(data=tips_df, x="total_bill", y="tip")

ax.set_title(
    "Analisa Hubungan Antara Total Tagihan dan Tip",
    fontsize=18,
    fontweight='bold',
    )
ax.set_xlabel(
    "Total Tagihan Pelanggan", fontsize=12
)
ax.set_ylabel(
    "Jumlah Tip", fontsize=12
)
mean_tip = tips_df['tip'].mean()
ax.axhline(mean_tip,color='red',linestyle='--', label=f'Rata rata tip: {mean_tip: 2.2f}')
plt.legend()
plt.show()
```
