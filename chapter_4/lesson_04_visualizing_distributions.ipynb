### 1. BoxPlot

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
```

```python
# Membuat Data
tips_df = sns.load_dataset('tips')
sns.set_theme(style='whitegrid')
```

```python
plt.figure(figsize=(8,4))
sns.boxplot(data=tips_df, x='day', y='total_bill', hue='sex')
plt.title("Distribusi Total Tagihan Perhari")
plt.xlabel('Hari')
plt.ylabel('Total Tagihan Ratus Rp')
plt.show()
```

### 2. ViolinPlot

```python
plt.figure(figsize=(8,4))
sns.violinplot(data=tips_df, x='day', y='total_bill', hue='sex')
plt.title("Distribusi Total Tagihan Perhari")
plt.xlabel('Hari')
plt.ylabel('Total Tagihan Ratus Rp')
plt.show()
```

### 3. StripPlot

```python
plt.figure(figsize=(8,4))
sns.stripplot(data=tips_df, x='smoker', y='tip')

plt.title("Sebaran tip: perokok vs non-perokok")
plt.show()

plt.figure(figsize=(8,4))
sns.stripplot(data=tips_df, x='smoker', y='tip', jitter=True)

plt.title("Sebaran tip: perokok vs non-perokok")
plt.show()
```

### 4. Menggabungkan Plot, violin dengan strip

```python
plt.figure(figsize=(8,4))

sns.violinplot(data=tips_df, x='day', y='total_bill', inner=None, color='lightgray')

sns.stripplot(data=tips_df, x='day', y='total_bill', jitter=True, hue='smoker')

plt.title("Distribusi Total Tagihan dengan Titik Data Individual")
plt.show()
```
